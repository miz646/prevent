```python
import pandas as pd

#df = pd.read_csv("dementia_analysis.csv")
#df = pd.read_csv("DLBvsCN_.csv")
df = pd.read_csv("ADvsCN_.csv")

df.head()
```

```python
df["Sex"] = df["Sex"].astype("float64")
df["SampleGroup"] = df["SampleGroup"].astype("float64")
df["Age"] = df["Age"].astype("float64")
```

```python
df1 = df.drop("SampleId",axis=1)
df1 = df1.drop("SampleGroup", axis =1)
df1.head()
```

```python
#説明変数を標準化
from sklearn.preprocessing import StandardScaler


# 標準化する列を選択
cols_to_scale = list(df1.columns)
#cols_to_scale.append("Age")
#cols_to_scale.append("Sex")


# StandardScalerをインスタンス化する
scaler = StandardScaler()

# 選択した列の値を標準化する
df1[cols_to_scale] = scaler.fit_transform(df1[cols_to_scale])

df1.head()
```

```python
#相互情報量、ランダムフォレスト、GBDTで、上位n位までの表示を指定
n = 50
```

```python
#ランダムフォレスト

from sklearn.ensemble import RandomForestClassifier

X = df1
y = df["SampleGroup"]

clf = RandomForestClassifier(n_estimators=10 , random_state=71)
clf.fit(X,y)
fi = clf.feature_importances_

import numpy as np

idx1 = np.argsort(fi)[::-1]
top_cols_rf = X.columns.values[idx1][:n]
top_importance = fi[idx1][:n]
#top_cols = X.columns.values[idx][30:41]
print(top_cols_rf)
#print(top_importance)
```

```python
#相互情報量

from sklearn.feature_selection import mutual_info_classif

X = df1
y = df["SampleGroup"]

mi = mutual_info_classif(X,y)

import numpy as np

idx = np.argsort(mi)[::-1]
top_cols_mi = X.columns.values[idx][:n]
#top_cols = X.columns.values[idx][30:41]
print(top_cols_mi)
```

```python
#XGBoost

import xgboost as xgb

X = df1
y = df["SampleGroup"]

dtrain = xgb.DMatrix(X, label = y)
params ={ 'objective' : 'binary:logistic', 'silent' : 1, 'random_state' : 71 }
#num_round = n
model = xgb.train(params, dtrain, n)

fscore = model.get_score(importance_type = 'total_gain')
print(fscore)
fscore = sorted([(k,v) for k,v in fscore.items()], key = lambda tpl: tpl[1], reverse = True)

res = fscore[:n]

#print(res)

data = res

top_cols_xgb = [item[0] for item in data if item[0].startswith('seq')]

print(top_cols_xgb)
print(len(res))
```

```python
mi_rf_xgb = set(top_cols_mi) & set(top_cols_rf) &set(top_cols_xgb)
mi_rf = set(top_cols_mi) & set(top_cols_rf)
rf_xgb = set(top_cols_rf) & set(top_cols_xgb)
mi_xgb = set(top_cols_mi) & set(top_cols_xgb)

print("mi,rf,xgbの各上位50位の全てに共通のseq.", mi_rf_xgb)
print("mi,rfの各上位50位で共通のseq." , mi_rf)
print("rf,xgbの各上位50位で共通のseq.", rf_xgb)
print("mi,xgbの各上位50位で共通のseq.", mi_xgb)
```
